```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
def scrape_reviews(pages=5):

    base_url = "https://www.trustpilot.com/review/shopee.com.my?page="

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    all_reviews = []

    for page in range(1, pages+1):

        url = base_url + str(page)
        response = requests.get(url, headers=headers)

        soup = BeautifulSoup(response.text, "html.parser")

        reviews = soup.find_all("article")

        for review in reviews:

            name = review.find("span", {"data-consumer-name": True})
            date = review.find("time")
            content = review.find("p")

            reviewer = name.text.strip() if name else ""
            review_date = date.text.strip() if date else ""
            review_text = content.text.strip() if content else ""

            all_reviews.append([reviewer, review_date, review_text])

        print(f"Page {page} scraped successfully")
        time.sleep(2)

    df = pd.DataFrame(
        all_reviews,
        columns=["Reviewer Name","Review Date","Review Content"]
    )

    return df
```

```python
df_reviews = scrape_reviews(pages=5)
df_reviews.head()
```

```python
df_reviews.to_csv("reviews.csv", index=False)
print("CSV saved!")
```
