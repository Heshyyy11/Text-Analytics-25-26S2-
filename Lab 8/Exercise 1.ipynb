TF-IDF with Preprocessing

```python
import numpy as np
import re
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from tabulate import tabulate
from collections import Counter
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Dataset
dataset = [
    "I love playing football on the weekends",
    "I enjoy hiking and camping in the mountains",
    "I like to read books and watch movies",
    "I prefer playing video games over sports",
    "I love listening to music and going to concerts"
]

# Preprocessing function
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    words = text.split()
    words = [w for w in words if w not in ENGLISH_STOP_WORDS]
    return " ".join(words)

# Apply preprocessing
cleaned_dataset = [preprocess(doc) for doc in dataset]

# TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(cleaned_dataset)

# Clustering
k = 2
km = KMeans(n_clusters=k, random_state=0)
y_pred = km.fit_predict(X)

# Output
table_data = [["Document", "Cluster"]]
table_data.extend([[doc, cluster] for doc, cluster in zip(dataset, y_pred)])
print(tabulate(table_data, headers="firstrow"))

# Purity
purity = max(Counter(y_pred).values()) / len(y_pred)
print("Purity:", purity)
```

Word2Vec with Preprocessing

```python
import numpy as np
import re
from sklearn.cluster import KMeans
from gensim.models import Word2Vec
from tabulate import tabulate
from collections import Counter
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Preprocessing
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    words = text.split()
    words = [w for w in words if w not in ENGLISH_STOP_WORDS]
    return words

# Apply preprocessing
tokenized_dataset = [preprocess(doc) for doc in dataset]

# Train Word2Vec
model = Word2Vec(sentences=tokenized_dataset, vector_size=100, window=5, min_count=1)

# Sentence vectors
X = np.array([
    np.mean([model.wv[word] for word in doc], axis=0)
    for doc in tokenized_dataset
])

# Clustering
k = 2
km = KMeans(n_clusters=k, random_state=0)
y_pred = km.fit_predict(X)

# Output
table_data = [["Document", "Cluster"]]
table_data.extend([[dataset[i], y_pred[i]] for i in range(len(dataset))])
print(tabulate(table_data, headers="firstrow"))

# Purity
purity = max(Counter(y_pred).values()) / len(y_pred)
print("Purity:", purity)
```

Yes, the purity increased after text preprocessing.
